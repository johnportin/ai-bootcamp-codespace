```python
from github import GithubRepositoryDataReader
import minsearch
```

```python
repo_owner = "DataTalksClub"
repo_name = "datatalksclub.github.io"
allowed_extensions = ["md", "mdx"]
filename_filter = lambda x: "_podcast" in x
# filename_filter = "tree/main/_podcast"
```

```python
reader = GithubRepositoryDataReader(repo_owner, repo_name, allowed_extensions)
```

```python
data = reader.read()
```

```python
for d in data:
    print(d.filename)
```

```python
i = 0
# for line in data[7].filename.split("\n"):
for d in data:
    # print(i, d.filename)
    if "ai-for-digital-health" in d.filename:
        print(i, d.filename)
    i += 1
```

```python
print(f"There are {len(data)} records in the podcasts dataset.")
```

```python
def paragraph_chunk(data, size, step):
    chunks = []

    for d in data:
        podcast_name = d.filename.split("/")[-1].replace(".md", "")
        paragraphs = [p.strip() for p in d.content.split("\n") if p.strip()]

        for i in range(0, len(paragraphs), step):
            text = " ".join(paragraphs[i:i+size])
            if not text:
                continue
            chunks.append({
                "podcast": podcast_name,
                "chunk": text
            })

    return chunks
```

```python
chunks = paragraph_chunk(data, 30, 15)
```

```python
for chunk in chunks:
    if "ai-for-digital-health" in chunk['podcast']:
        print(chunk['chunk'].replace("\ \ ", " "))
```

```python
print(f"There are {len(chunks)} chunks in the processed podcasts dataset.")
```

```python
index = minsearch.Index(
    text_fields=["chunk"],
    keyword_fields=["podcast"]
)
index.fit(chunks)
```

```python
query = "how do I make money with AI?"
results = index.search(query)
```
